```python
import pandas as pd

reactores = pd.read_csv('/content/IAEA_Reactores_Nucleares_Investigacion.csv')
```

```python
# Ubicaciones
ubicaciones = reactores[['Pais', 'Ciudad']]
ubicaciones = ubicaciones.fillna('No registrado')
ubicaciones = ubicaciones.drop_duplicates()
ubicaciones.columns = ['nombre_pais', 'nombre_ciudad']
ubicaciones.to_csv('ubicaciones.csv', sep=';', index=False)
ubicaciones
```

```python
# Tipo de reactores

tipo_reactores = reactores['Tipo'].unique()
tipo_reactores = pd.DataFrame(tipo_reactores, columns=['nombre'])
tipo_reactores = tipo_reactores.fillna('No registrado')
ciudades = ciudades.drop_duplicates(subset='nombre')
tipo_reactores.to_csv('tipo_reactores.csv', sep=';', index=False)
tipo_reactores
```
